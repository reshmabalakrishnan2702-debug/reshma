```python
def amt_calculator(amt,withdraw):
  amount=amt-withdraw if amt-withdraw>=2000 else 0
  return amount

balance=eval(input("enter the balance amount:"))
withdrawal_amount=int(input("enter the withdrawal amount"))
amount=amt_calculator(balance,withdrawal_amount)
if amount!=0:
  print(f"transaction success:/n Balance amout:{amount}")
else:
  print("sorry insuffient amount")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target_names[iris_data.target]


# Pairplot with numeric columns only
numeric_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
sns.pairplot(iris[numeric_cols + ['species']], hue='species', diag_kind='hist')
plt.show()

# Boxplot showing petal_width distribution for each species
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal width (cm)', data=iris)
plt.title('Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

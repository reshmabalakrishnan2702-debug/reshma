```python
"""prgm for finding sum of even number from 1 to n where negative number case not considered"""
flag =True
while flag:
  limit=int(input("enter the limit upto which you want to find even sum"))
  if limit<0:
    print("negative number not accepted")
    exit()
  else:
    break
sum=0
for i in range(1,limit+1):
  if i%2==0:
    sum+=i
print(f"sum of even number from 1 to {limit} is : {sum}")
```

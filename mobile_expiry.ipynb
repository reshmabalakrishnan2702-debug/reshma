```python
from datetime import datetime
def date_difference(recharge_date,today_date):
  diff_days=today_date-recharge_date
  return  diff_days.days

recharge_date = input("Enter the recharge date (YYYY-MM-DD): ")
today_date = input("Enter  todays date (YYYY-MM-DD): ")
try:
        date1= datetime.strptime(recharge_date, "%Y-%m-%d")
        date2 = datetime.strptime(today_date, "%Y-%m-%d")
except ValueError:
        print("Invalid date format. Please use YYYY-MM-DD.")
        exit(0)
value=date_difference(date1,date2)
if value>0:
  print(f"days left for charging:{28-value}")
else:
  print("recharge")
```

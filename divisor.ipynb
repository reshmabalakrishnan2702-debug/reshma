```python
range_limit=int(input("enter the limit    "))
divisor=int(input("enter the divisor to check    "))
elements=[]
count=0
for i in range(1,range_limit+1):
    if i%divisor==0:
      elements.append(i)
      count+=1
print(f"count of number divisible by {divisor} is :{count}")
print(tuple(elements))
```

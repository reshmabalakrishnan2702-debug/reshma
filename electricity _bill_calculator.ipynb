```python

def bill_calculator(units):
  if units<=100:
    total_amount =units*5
  elif units>=200:
    total_amount = 100*5+(units-100)*7
  else:
    total_amount =100*12+(units-200)*10
  return total_amount

units=float(input("enter the units"))
print(f"Bill amount: {bill_calculator(units)}\n please pay")



```

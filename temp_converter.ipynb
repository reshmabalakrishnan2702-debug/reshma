


# function for conversion

```python
def temp_conversion(temp,unit):
  value=((temp*9)/5)+32 if unit=='c' else (temp-32)*5/9
  if unit=='c':
   new_unit='f'
  else:
   new_unit='c'
  return value,new_unit

temperature=float(input("enter a value to convert"))
units=input("enter unit c/f")
if units not in 'cf':
  print("wrong input")
  exit(0)
else:
  temp,newunit=temp_conversion(temperature,units)
print(f"temperature in {newunit} is {temp}")


```

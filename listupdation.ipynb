```python
names = ['reshma','roshni','arya','archana','aditya','bharat']
print(f" list of names: {names}")


def name_removal(names):
  name_to_remove = input("Enter the name you want to remove: ")
  if name_to_remove in names:
      position=names.index(name_to_remove)
      names.remove(name_to_remove)
      print(f"List after removing '{name_to_remove}': {names}")
  else:
      print(f"{name_to_remove} not found in list.")
  for i in range(position,len(names)-1):
      names[i]=names[i+1]

def insert_location(names):
  name_to_insert = input("Enter name to insert: ")
  position_to_insert = int(input("Enter position to insert: "))
  names.insert(position_to_insert, name_to_insert)
  print(f"List after inserting '{name_to_insert}' at position {position_to_insert}: {names}")

def display(names):
  print("\nBooked:")
  for name in names:
    print(name)


while True:
  print("\n--- List Operations Menu ---")
  print("1. remove item from the list")
  print("2. insert name  by position")
  print("3. Display list")
  print("4. Exit")
  choice = input("Enter your choice (1-4): ")

  if choice == '1':
    name_removal(names)
  elif choice == '2':
    insert_location(names)
  elif choice == '3':
    display(names)
  elif choice == '4':
    print("Exiting program. Goodbye!")
    break
  else:
    print("Invalid choice. Please enter a number between 1 and 5.")

```

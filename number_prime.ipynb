```python
"""prime number finder using function"""
def prime(num):
  counter =0
  for i in range(1,num+1):
    if num % i == 0:
      counter +=1
  if (num == 0) or (num == 1) or (counter >2):
    return False
  else:
    return True

number=int(input("enter a number to check prime or not  :"))
output=prime(number)
if output==True:
  print(f"{number} is prime")
else:
  print(f"{number} is not prime")


```
